```python
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("/content/gdrive/MyDrive/Colab Notebooks/botol.jpg")
cg = cv2.imread("/content/gdrive/MyDrive/Colab Notebooks/botol.jpg")
baris,kolom = img.shape[:2]
```

```python
#Menampilkan histogram
fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].imshow(img)
axs[1].hist(img.ravel(),256,[0,256])
plt.show()
```

```python
#Meningkatkan Kecerahan g(x,y) = f(x,y) + b
beta = 19
for x in range(baris) :
  for y in range(kolom):
    fyx = img[x,y] + beta
    cg[x,y] = fyx

fig, axs = plt.subplots(1,2,figsize=(15,5))
axs[0].imshow(cg)
axs[1].hist(cg.ravel(),256,[0,256])
plt.show()
```

```python
#Meregangkan Kontras g(x,y)=a*f(x,y)
alpa = 76
for x in range(baris) :
  for y in range(kolom):
    fyx = img[x,y] * alpa
    cg[x,y] = fyx

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])
axs[1,0].imshow(cg)
axs[1,1].hist(cg.ravel(),256,[0,256])
plt.show()
```

```python
#Kombinasi Kecerahan dan Kontras g(x,y)=a.f(x,y)+b
alpa = 76
beta = 19
for x in range(baris) :
    for y in range(kolom) :
        fyx = alpa * img[x,y] + beta
        cg[x,y] = fyx

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])
axs[1,0].imshow(cg)
axs[1,1].hist(cg.ravel(),256,[0,256])
plt.show()
```

```python
#Membalik Citra g(x,y) = 255-f(x,y)
for x in range(baris) :
  for y in range(kolom) :
        fyx = 255 - img[x,y]
        cg[x,y] = fyx

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])
axs[1,0].imshow(cg)
axs[1,1].hist(cg.ravel(),256,[0,256])
plt.show()
```

```python
#Memotong Aras Keabuan
f1 = 30
f2 = 200
for x in range(baris) :
    for y in range(kolom) :
        fyx = img[x,y]
        for z in range(3) : 
            if fyx[z] <= f1 :
                cg[x,y][z] = 0
            elif fyx[z] >= f2 :
                cg[x,y][z] = 255
            else :
                cg[x,y][z] = fyx[z]

fig, axs = plt.subplots(2,2,figsize=(15,5))
axs[0,0].imshow(img)
axs[0,1].hist(img.ravel(),256,[0,256])
axs[1,0].imshow(cg)
axs[1,1].hist(cg.ravel(),256,[0,256])
plt.show()
```
